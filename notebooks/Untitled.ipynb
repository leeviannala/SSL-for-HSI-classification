```python
def make_string(ds_list, val, val_string):
    ret = f'{val} {val_string}'
    summa = 0
    for ds in ds_list:
        summa += len(ds[ds==val])
        ret += f' {len(ds[ds==val])}'
    ret += f' {summa}'
    return ret
print('Class Kasos Tilos Anafi Atokos Donousa Sum')
for i, val in enumerate(['Undefined',
                        'High intensity developed', 
                        'Med-low intensity developed',
                        'Deciduous, Evergreen, mixed forest',
                        'shrubland',
                        'Grassland-Pasture',
                        'Bareland',
                        'water',
                        'corn',
                        'cotton',
                        'cereals',
                        'almonds',
                        'grass fodders',
                        'vinewards-grapes',
                        'walnuts',
                        'pistachios',
                        'citrus',
                        'fallow']):
    print(make_string([kasos, tilos, anafi, atokos, donousa], i, val))
```
